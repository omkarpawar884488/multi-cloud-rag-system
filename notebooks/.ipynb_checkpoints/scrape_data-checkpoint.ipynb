```python
import re, os, json
import numpy as np
import pandas as pd
import fitz
from PIL import Image, ImageOps
import pytesseract


pytesseract.pytesseract.tesseract_cmd = r"C:\Program Files\Tesseract-OCR\tesseract.exe"

PDF_PATH = r"E://CV//Internship//biodata//melava_pdf//melava bookC2C.pdf"
START_PAGE = 14
END_PAGE = 42
DPI = 300
OCR_LANG = "mar+eng"
OUTPUT_XLSX = "profiles_page14_to_42.xlsx"
```

```python
def render_page_to_pil(doc, page_index0, dpi=300):
    page = doc.load_page(page_index0)
    zoom = dpi / 72.0
    mat = fitz.Matrix(zoom, zoom)
    pix = page.get_pixmap(matrix=mat, alpha=False)
    return Image.frombytes("RGB", (pix.width, pix.height), pix.samples)

def preprocess(img: Image.Image) -> Image.Image:
    # Light preprocessing helps OCR stability
    g = ImageOps.grayscale(img)
    g = ImageOps.autocontrast(g)
    return g

def ocr_text(img):
    img = preprocess(img)
    # PSM 6 = assume a block of text; often best for these cards
    config = r"--oem 3 --psm 6"
    return pytesseract.image_to_string(img, lang=OCR_LANG, config=config)
```

```python
LABELS = {
    "name": ["नाव", "Name"],
    "dob": ["जन्म", "DOB", "Date of Birth"],
    "education": ["शिक्षण", "Education"],
    "job": ["नोकरी", "Occupation", "Job"],
    "income": ["उत्पन्न", "Income"],
    "height": ["उंची", "Height"],
    "weight": ["वजन", "Weight"],
    "blood_group": ["रक्तगट", "Blood"],
    "complexion": ["वर्ण", "Complexion"],
    "caste": ["जात", "Caste"],
    "expectation": ["अपेक्षा", "Expectation"],
    "guardian_parent": ["पालक", "Parent", "Guardian"],
    "address": ["पत्ता", "Address"],
    "phone": ["दूरध्वनी", "फोन", "मोबाईल", "Mobile", "Phone"],
}

def clean_text(t: str) -> str:
    t = t.replace("\r", "\n")
    t = re.sub(r"\n{2,}", "\n", t)
    t = re.sub(r"[ \t]+", " ", t)
    return t.strip()

def is_label_line(line: str):
    s = line.strip()
    for field, variants in LABELS.items():
        for v in variants:
            # label at line start OR "label :" pattern
            if re.match(rf"^{re.escape(v)}(\s*[:\-]|\b)", s, flags=re.IGNORECASE):
                return field, v
    return None, None

def extract_profile_no(text: str):
    # also catches Marathi digits
    m = re.search(r"(क्र|क्)\s*[:.\-]?\s*([0-9०-९]+)", text)
    return m.group(2) if m else np.nan

def parse_candidate(text: str) -> dict:
    t = clean_text(text)
    lines = [ln.strip() for ln in t.split("\n") if ln.strip()]

    out = {k: np.nan for k in ["profile_no"] + list(LABELS.keys())}
    out["raw_text"] = t
    out["profile_no"] = extract_profile_no(t)

    current_field = None
    buffer = []

    def flush():
        nonlocal current_field, buffer
        if current_field is None:
            buffer = []
            return
        val = " ".join(buffer).strip()
        out[current_field] = val if val else np.nan
        buffer = []

    for ln in lines:
        field, matched_label = is_label_line(ln)

        if field is not None:
            # finish previous field
            flush()
            current_field = field

            # take remainder after label / colon if present
            remainder = re.sub(rf"^{re.escape(matched_label)}\s*[:\-]?\s*", "", ln, flags=re.IGNORECASE).strip()
            if remainder:
                buffer.append(remainder)
            continue

        # normal line -> belongs to the current field if one is active
        if current_field is not None:
            buffer.append(ln)

    flush()
    return out
```

```python
def extract_profiles_to_excel():
    rows = []

    with fitz.open(PDF_PATH) as doc:
        for pdf_page_num in range(START_PAGE, END_PAGE + 1):
            img = render_page_to_pil(doc, pdf_page_num - 1, dpi=DPI)
            w, h = img.size

            # Split into left and right candidate cards
            halves = [
                ("left",  (0, 0, w // 2, h)),
                ("right", (w // 2, 0, w, h)),
            ]

            for side, box in halves:
                crop = img.crop(box)
                text = ocr_text(crop)

                # Skip if OCR is empty/noisy
                if len(clean_text(text)) < 30:
                    continue

                d = parse_candidate(text)
                d["source_pdf_page"] = pdf_page_num
                d["side"] = side
                rows.append(d)

    df = pd.DataFrame(rows)

    cols = [
        "profile_no","name","dob","education","job","income","height","weight",
        "blood_group","complexion","caste","expectation","guardian_parent",
        "address","phone","source_pdf_page","side","raw_text"
    ]
    for c in cols:
        if c not in df.columns:
            df[c] = np.nan
    df = df[cols].replace(r"^\s*$", np.nan, regex=True)

    df.to_excel(OUTPUT_XLSX, index=False)
    print(f"Saved: {OUTPUT_XLSX} | rows={len(df)}")

extract_profiles_to_excel()
```
