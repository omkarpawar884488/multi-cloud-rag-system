```python
import requests
import bs4
import yaml
from pathlib import Path
```

```python
# if notebook is in cloud-rag-scraper/notebooks
PROJECT_ROOT = Path.cwd().parent  # parent of notebooks/ = project root
CONFIG_PATH = PROJECT_ROOT / "config" / "sources.yaml"

print("PROJECT_ROOT:", PROJECT_ROOT)
print("CONFIG_PATH:", CONFIG_PATH)
print("Exists?:", CONFIG_PATH.exists())
```

```python
def load_sources():
    with open(CONFIG_PATH, "r", encoding="utf-8") as f:
        return yaml.safe_load(f)

cfg = load_sources()
for provider, data in cfg["providers"].items():
    print(f"Provider: {provider}")
    for section, urls in data.get("web_roots", {}).items():
        print(f"  {section}: {len(urls)} URLs")
    print(f"  GitHub repos: {len(data.get('github_repos', []))}")
```
