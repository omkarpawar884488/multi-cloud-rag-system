```python
from pathlib import Path
import json, hashlib

BASE_DIR = Path(r"E:/GenAi/project/Rag_cloud_platforms")
registry_dir = BASE_DIR / "data/processed/registry"
registry_dir.mkdir(parents=True, exist_ok=True)
```

```python
INPUT_DIRS = [
    BASE_DIR / "data/processed/docs/github_normalized",
    BASE_DIR / "data/processed/docs/azure_web",
    BASE_DIR / "data/processed/docs/aws_web",
    BASE_DIR / "data/processed/docs/gcp_web",
]

out_path = registry_dir / "docs_registry.jsonl"

def clean_text(t: str) -> str:
    return " ".join((t or "").split())

def sha(text: str) -> str:
    return hashlib.sha256(text.encode("utf-8")).hexdigest()

count = 0
with out_path.open("w", encoding="utf-8") as out:
    for folder in INPUT_DIRS:
        if not folder.exists():
            print("Missing:", folder)
            continue
        for fp in folder.glob("*.json"):
            doc = json.loads(fp.read_text(encoding="utf-8"))
            text = doc.get("body_markdown") or doc.get("body_text") or ""
            text = clean_text(text)
            if len(text) < 300:
                continue

            unified = {
                "doc_id": doc.get("doc_id") or doc.get("url") or str(fp),
                "provider": doc.get("provider", "unknown"),
                "source_type": doc.get("source_type", "unknown"),
                "category": doc.get("category", "unknown"),
                "title": clean_text(doc.get("title", "")),
                "text": text,
                "url": doc.get("url"),
                "metadata": doc.get("metadata", {}),
                "text_hash": sha(text),
            }
            out.write(json.dumps(unified, ensure_ascii=False) + "\n")
            count += 1

print("Wrote:", out_path, "docs:", count)
```

```python
with out_path.open("r", encoding="utf-8") as f:
    first = json.loads(next(f))
first.keys(), first["provider"], first["title"][:100]
```
