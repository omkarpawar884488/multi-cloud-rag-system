```python
import re
import numpy as np
import pandas as pd

inp = r"E://GenAi//project//Rag_cloud_platforms//notebooks//profiles_page14_to_42.xlsx"   # your current file
df = pd.read_excel(inp)

df.columns
```

```python
def split_into_candidates(raw: str):
    """
    Split a raw_text string into multiple candidate blocks.
    A new candidate starts at a line containing 'नाव' (usually 'नाव :').
    Returns a list of blocks (strings). If no split possible, returns [raw].
    """
    if not isinstance(raw, str) or not raw.strip():
        return []

    text = raw.replace("\r", "\n")
    text = re.sub(r"\n{2,}", "\n", text).strip()

    # Find 'नाव' occurrences that look like a candidate header.
    # We look for 'नाव' on a new line (or start) and followed by ':' or '-'
    # allowing a few junk chars before नाव (OCR sometimes adds 'Fay ' etc.)
    pattern = re.compile(r"(?:^|\n)\s*.{0,6}?\bनाव\b\s*[:\-]", flags=re.UNICODE)

    matches = list(pattern.finditer(text))
    if not matches:
        return [text]

    starts = [m.start() for m in matches]
    blocks = []
    for i, s in enumerate(starts):
        e = starts[i+1] if i+1 < len(starts) else len(text)
        block = text[s:e].strip()
        if block:
            blocks.append(block)

    return blocks


def extract_name(text: str):
    """
    Extract candidate name from 'नाव' line.
    Handles OCR noise like 'Fay नाव :' etc.
    """
    if not isinstance(text, str):
        return np.nan

    t = text.replace("\r", "\n")

    # Match lines containing 'नाव : <name>'
    m = re.search(
        r"(?:^|\n)\s*.{0,6}?\bनाव\b\s*[:\-]?\s*([^\n]+)",
        t,
        flags=re.UNICODE
    )

    if not m:
        return np.nan

    name = m.group(1).strip()

    # Remove trailing punctuation
    name = re.sub(r"[.\s]+$", "", name)

    return name if name else np.nan

```

```python
# Keep only the columns you mentioned (add others if you want)
base_cols = ["source_pdf_page", "side", "raw_text"]
work = df[base_cols].copy()

# Create list of candidate blocks
work["candidate_blocks"] = work["raw_text"].apply(split_into_candidates)

# Explode into multiple rows
work = work.explode("candidate_blocks", ignore_index=True)

# Rename to raw_text_individual
work = work.rename(columns={"candidate_blocks": "raw_text_individual"})

# Drop empty
work["raw_text_individual"] = work["raw_text_individual"].astype(str).str.strip()
work = work[work["raw_text_individual"].str.len() > 0].copy()

work.head(10)
```

```python
def extract_first_line_value(text: str, label_regex: str):
    """
    Extracts the value after a label from the same line.
    Example: 'शिक्षण : MBA' -> 'MBA'
    Returns NaN if not found.
    """
    m = re.search(label_regex, text, flags=re.IGNORECASE)
    if not m:
        return np.nan
    val = m.group(1).strip()
    return val if val else np.nan

def remove_label_line(text: str, label_line_regex: str):
    """Remove the full line that contains a label (used for other_info)."""
    return re.sub(label_line_regex, "", text, flags=re.IGNORECASE)

def parse_fields(block: str):
    t = block.replace("\r", "\n")
    t = re.sub(r"\n{2,}", "\n", t).strip()

    # ---- NAME ----
    name = extract_name(t)

    # ---- DOB ----
    dob = extract_first_line_value(
        t,
        r"जन्म\s*ता\.?\s*[:\-]?\s*([^\n]+)"
    )

    # ---- EDUCATION ----
    education = extract_first_line_value(
        t,
        r"शिक्षण\s*[:\-]?\s*([^\n]+)"
    )

    # ---- OTHER INFO ----
    other = t
    other = remove_label_line(other, r"^.*नाव\b.*$\n?")
    other = remove_label_line(other, r"^.*जन्म\s*ता\.?.*$\n?")
    other = remove_label_line(other, r"^.*शिक्षण.*$\n?")

    other = re.sub(r"\n{2,}", "\n", other).strip()
    other = other if other else np.nan

    return name, dob, education, other


work[["नाव", "dob", "education", "other_info"]] = work["raw_text_individual"].apply(
    lambda x: pd.Series(parse_fields(x))
)

work.head(10)
```

```python
final_cols = [
    "नाव",
    "dob",
    "education",
    "other_info",
    "source_pdf_page",
    "side",
]

# Keep only these columns and in this order
final_df = work[final_cols].copy()

out = r"profiles_split_and_parsed.xlsx"
final_df.to_excel(out, index=False)

out
```
