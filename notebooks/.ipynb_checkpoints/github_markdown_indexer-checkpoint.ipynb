```python
import os
from pathlib import Path
import json
import hashlib
```

```python
# --- CONFIG (you can change these easily in a notebook) ---
GITHUB_ROOT = Path("data/raw/github")  # root folder that contains all the cloned repos
OUTPUT_PATH = Path("data/processed/docs/github_index.jsonl")


def iter_markdown_files(root: Path):
    """
    Yield all markdown files under `root`.
    """
    root = Path(root)
    for dirpath, _, filenames in os.walk(root):
        for fname in filenames:
            if fname.lower().endswith((".md", ".markdown")):
                yield Path(dirpath) / fname


def detect_provider(path: Path) -> str:
    """
    Infer cloud provider (azure, gcp, aws) from file path.
    Falls back to 'unknown' if nothing matches.
    """
    p = str(path).lower()

    if "azure" in p:
        return "azure"
    if "gcp" in p or "googlecloudplatform" in p:
        return "gcp"
    if "aws" in p or "amazon" in p:
        return "aws"
    return "unknown"


def make_doc_id(path: Path, root: Path) -> str:
    """
    Make a stable doc id from path relative to `root`.
    """
    rel = Path(path).relative_to(root)
    return rel.as_posix()  # POSIX-style (forward slashes) for consistency


def file_hash(path: Path, chunk_size: int = 8192) -> str:
    """
    SHA-256 hash of file contents (streamed, so big files are ok).
    """
    h = hashlib.sha256()
    with open(path, "rb") as f:
        for chunk in iter(lambda: f.read(chunk_size), b""):
            h.update(chunk)
    return h.hexdigest()


def build_index(root: Path = GITHUB_ROOT,
                output_path: Path = OUTPUT_PATH,
                overwrite: bool = True) -> None:
    """
    Build a JSONL index of all markdown files under `root`.
    Designed to be called directly from a Jupyter cell.
    """
    root = Path(root)
    output_path = Path(output_path)

    output_path.parent.mkdir(parents=True, exist_ok=True)

    if output_path.exists() and not overwrite:
        raise FileExistsError(f"{output_path} already exists. "
                              f"Set overwrite=True to overwrite it.")

    count = 0
    with open(output_path, "w", encoding="utf-8") as out:
        for md_path in iter_markdown_files(root):
            doc = {
                "doc_id": make_doc_id(md_path, root),
                "provider": detect_provider(md_path),
                "source_type": "github",
                "repo_rel_path": str(md_path.relative_to(root)),
                "abs_path": str(md_path.resolve()),
                "content_hash": file_hash(md_path),
            }
            out.write(json.dumps(doc, ensure_ascii=False) + "\n")
            count += 1

    print(f"Wrote {count} documents to {output_path}")
```

```python
build_index(root=Path("E:/GenAi/project/Rag_cloud_platforms/data/raw/github"),
            output_path=Path("E:/GenAi/project/Rag_cloud_platforms/data/processed/docs/github_index.jsonl"))
```
