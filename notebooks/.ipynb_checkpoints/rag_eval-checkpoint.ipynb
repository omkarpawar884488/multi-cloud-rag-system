```python
def topic_cov_score_from_out(out: dict) -> float:
    tc = out.get("topic_coverage") or {}
    req = tc.get("requested", [])
    miss = tc.get("topic_missing", [])
    if not req:
        return 1.0
    return max(0.0, 1.0 - (len(miss) / len(req)))



def validate_grounding(out: dict) -> dict:
    answer = (out.get("answer") or "").lower()
    grounding = out.get("grounding") or {}
    missing = set(grounding.get("missing", []))

    flags = []

    # Allowed "missing" phrases (mentions are OK if they are clearly stating missing context)
    missing_ok_patterns = [
        r"information for aws .* missing",
        r"details for aws .* missing",
        r"context .* (does not|doesn't) provide .* aws",
        r"aws .* not available in the provided context",
        r"missing .* aws",
        r"information for gcp .* missing",
        r"details for gcp .* missing",
        r"context .* (does not|doesn't) provide .* gcp",
        r"gcp .* not available in the provided context",
        r"missing .* gcp",
    ]

    def is_only_missing_statement(provider: str) -> bool:
        if provider not in answer:
            return True
        return any(re.search(pat, answer) for pat in missing_ok_patterns)

    # Hallucination patterns: provider + strong verb/claim (adjust as you like)
    aws_claim_patterns = [
        r"aws landing zone (provides|is|does|automates|creates|enforces)",
        r"aws (uses|offers|supports|implements)",
    ]
    gcp_claim_patterns = [
        r"gcp landing zone (provides|is|does|uses|implements)",
        r"gcp (uses|offers|supports|implements)",
    ]

    if "aws" in missing:
        if any(re.search(p, answer) for p in aws_claim_patterns) and not is_only_missing_statement("aws"):
            flags.append("AWS factual claims present but AWS context missing")

    if "gcp" in missing:
        if any(re.search(p, answer) for p in gcp_claim_patterns) and not is_only_missing_statement("gcp"):
            flags.append("GCP factual claims present but GCP context missing")

    # Citation hygiene (still rough)
    if "[" not in out.get("answer", ""):
        flags.append("No citations found in answer")

    return {
        "missing_providers": sorted(missing),
        "flags": flags,
        "ok": len(flags) == 0
    }


def score_answer(out: dict) -> dict:
    answer = (out.get("answer") or "")
    ans_l = answer.lower()

    grounding = out.get("grounding") or {}
    requested = set(grounding.get("requested", []))
    present = set(grounding.get("present", []))
    missing = set(grounding.get("missing", []))

    # A) Provider coverage (same as before)
    if requested:
        provider_coverage = len(present.intersection(requested)) / len(requested)
    else:
        provider_coverage = 1.0

    # NEW: A2) Topic coverage (Step 3)
    topic_coverage = topic_cov_score_from_out(out)

    # B) Grounding (same logic as before)
    missing_ok = re.search(r"(missing|not available|not provided).*(aws|gcp|azure)", ans_l) is not None

    aws_claim = re.search(r"aws landing zone (provides|does|automates|enforces|creates|is)", ans_l) is not None
    azure_claim = re.search(r"azure landing zone (provides|does|implements|uses|is)", ans_l) is not None
    gcp_claim = re.search(r"gcp landing zone (provides|does|implements|uses|is)", ans_l) is not None

    grounding_flags = []
    if "aws" in missing and aws_claim and not missing_ok:
        grounding_flags.append("AWS claims without AWS docs")
    if "azure" in missing and azure_claim and not missing_ok:
        grounding_flags.append("Azure claims without Azure docs")
    if "gcp" in missing and gcp_claim and not missing_ok:
        grounding_flags.append("GCP claims without GCP docs")

    grounding_score = 1.0 if not grounding_flags else 0.0

    # C) Citation score (same as before)
    citations = re.findall(r"\[\d+\]", answer)
    citation_score = 1.0 if len(citations) >= 2 else (0.5 if len(citations) == 1 else 0.0)

    # ✅ Final weighted score (Step 3 applied)
    # Recommended: topic coverage should matter at least as much as provider coverage.
    final = (
        0.30 * provider_coverage +
        0.40 * topic_coverage +
        0.20 * grounding_score +
        0.10 * citation_score
    )

    return {
        "provider_coverage": round(provider_coverage, 3),
        "topic_coverage": round(topic_coverage, 3),
        "grounding_score": round(grounding_score, 3),
        "citation_score": round(citation_score, 3),
        "final_score": round(final, 3),
        "grounding_flags": grounding_flags,
        "requested": sorted(requested),
        "present": sorted(present),
        "missing": sorted(missing),
        "topic_missing": (out.get("topic_coverage") or {}).get("topic_missing", []),
        "num_sources": len(out.get("sources", [])),
    }
```

```python
test_questions = [
    {"id": "Q001", "category": "Architecture", "question": "Design an AWS landing zone for dev/test/prod with guardrails and logging."},
    {"id": "Q002", "category": "Security", "question": "Explain IAM differences across AWS/Azure/GCP with examples."},
]

run_id = datetime.now().strftime("%Y%m%d_%H%M%S")
rows = []

for t in test_questions:
    qid = t["id"]
    cat = t["category"]
    question = t["question"]

    out = chain.invoke(question)

    # your checks
    grounding_check = validate_grounding(out)
    score = score_answer(out)

    # extract what your chain returns
    retrieval = out.get("retrieval") or out.get("retrieval_diag") or {}
    semantic = out.get("semantic_grounding") or out.get("semantic_diag") or {}

    rows.append({
        "run_id": run_id,
        "id": qid,
        "category": cat,
        "question": question,
        "answer": out.get("answer", ""),
        "num_sources": len(out.get("sources", [])),
        "sources": json.dumps(out.get("sources", []), ensure_ascii=False),
        "grounding": json.dumps(out.get("grounding", {}), ensure_ascii=False),
        "topic_coverage": json.dumps(out.get("topic_coverage", {}), ensure_ascii=False),

        # NEW diagnostics from updated rag_lcel
        "retrieval": json.dumps(retrieval, ensure_ascii=False),
        "semantic_grounding": json.dumps(semantic, ensure_ascii=False),

        # evaluation
        "grounding_flags": json.dumps(grounding_check.get("flags", []), ensure_ascii=False),
        "provider_coverage": score["provider_coverage"],
        "topic_coverage_score": score["topic_coverage"],
        "grounding_score": score["grounding_score"],
        "citation_score": score["citation_score"],
        "final_score": score["final_score"],
    })

df = pd.DataFrame(rows)
df


df = pd.json_normalize(rows)
#display(df[["id","category","score.final_score","score.provider_coverage","score.topic_coverage","score.grounding_score","score.citation_score"]])
```

```python
out_dir = PROJECT_ROOT / "rag_eval_runs"
out_dir.mkdir(parents=True, exist_ok=True)

jsonl_path = out_dir / f"rag_eval_{run_id}.jsonl"
csv_path = out_dir / f"rag_eval_{run_id}.csv"

with open(jsonl_path, "w", encoding="utf-8") as f:
    for r in rows:
        f.write(json.dumps(r, ensure_ascii=False) + "\n")

df.to_csv(csv_path, index=False)

print("Saved JSONL:", jsonl_path)
print("Saved CSV  :", csv_path)
```
