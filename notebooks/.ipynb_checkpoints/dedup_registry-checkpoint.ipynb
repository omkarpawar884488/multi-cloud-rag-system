```python
from pathlib import Path
import json
```

```python
# --- 1) Set BASE_DIR manually ---
# Option A: current working directory
# BASE_DIR = Path.cwd()

# Option B: explicitly set your project root
BASE_DIR = Path(r"E:\GenAi\project\Rag_cloud_platforms")

IN_PATH  = BASE_DIR / "data" / "processed" / "registry" / "docs_registry.jsonl"
OUT_PATH = BASE_DIR / "data" / "processed" / "registry" / "docs_registry_dedup.jsonl"


def dedup_registry(in_path: Path, out_path: Path, hash_key: str = "text_hash", verbose: bool = True):
    out_path.parent.mkdir(parents=True, exist_ok=True)

    seen = set()
    kept = 0
    total = 0
    skipped_no_hash = 0
    skipped_dupe = 0

    with in_path.open("r", encoding="utf-8") as f_in, out_path.open("w", encoding="utf-8") as f_out:
        for line in f_in:
            line = line.strip()
            if not line:
                continue

            total += 1
            doc = json.loads(line)

            h = doc.get(hash_key)
            if not h:
                skipped_no_hash += 1
                continue

            if h in seen:
                skipped_dupe += 1
                continue

            seen.add(h)
            f_out.write(json.dumps(doc, ensure_ascii=False) + "\n")
            kept += 1

    if verbose:
        print(f"Dedup done: {kept}/{total} docs kept → {out_path}")
        print(f"Skipped: {skipped_dupe} duplicates, {skipped_no_hash} missing {hash_key}")

    return {
        "total": total,
        "kept": kept,
        "skipped_duplicates": skipped_dupe,
        "skipped_no_hash": skipped_no_hash,
        "out_path": str(out_path),
    }


# --- 2) Run it ---
stats = dedup_registry(IN_PATH, OUT_PATH)
stats
```
