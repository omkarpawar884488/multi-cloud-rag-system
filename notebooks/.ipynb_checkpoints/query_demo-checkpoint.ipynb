```python
import chromadb
from chromadb.config import Settings
from sentence_transformers import SentenceTransformer
from pathlib import Path
import os
os.environ["CHROMA_ANONYMIZED_TELEMETRY"] = "FALSE"
```

```python
BASE_DIR = Path(r"E:/GenAi/project/Rag_cloud_platforms")
CHROMA_DIR = BASE_DIR / "data/vector_db/chroma"
```

```python
client = chromadb.PersistentClient(
    path=str(CHROMA_DIR),
    settings=Settings(
        anonymized_telemetry=False
    )
)

collection = client.get_collection("cloud_rag_chunks")
print("Count:", collection.count())
```

```python
# Load model
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")

# Query
query = "Explain AWS Well-Architected cost optimization pillar in simple terms"
q_emb = model.encode([query]).tolist()[0]

# Query Chroma
res = collection.query(
    query_embeddings=[q_emb],
    n_results=5,
    where={"provider": "aws"}  # optional filter
)

print("QUERY:", query)

for i, (doc, meta) in enumerate(
    zip(res["documents"][0], res["metadatas"][0]), 1
):
    print("\n---", i, "---")
    print("Title:", meta.get("title"))
    print("URL:", meta.get("url"))
    print(doc[:600], "...")
```
